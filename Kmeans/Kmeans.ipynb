```python
#import required packages
import numpy as np
from sklearn import datasets
from matplotlib import pyplot as plt
```

```python
#importing a dataset from sklearn and creating an array out of it
iris = datasets.load_iris()
X=iris.data
X.shape
#entering number of clusters we want
k=4
```

```python
#for our centres we take random rows from our dataset initially
a=np.random.choice(X.shape[0], size=k, replace=False)
centers = X[a, :]
print(centers)
```

```python
#plotting our initial output for initial visualization
plt.scatter(X[:,0],X[:,1],s=50)
plt.scatter(centers[:,0],centers[:,1],marker='*',c='g',s=50)
```

```python
#main kmeans algo
#where the magic happens!!
centers_new=centers
centers_old=np.zeros(centers.shape)
e=np.linalg.norm(centers_old-centers_new)
n=X.shape[0]
cluster=np.zeros(n)
distance = np.zeros((n,k))
while e!=0:
    for i in range(k):
        distance[:,i]=np.linalg.norm(X-centers[i],axis=1)
    cluster= np.argmin(distance, axis = 1)
    centers_old=centers_new
    for i in range(k):
        centers_new[i]=np.mean(X[cluster==i],axis=0)
    e=np.linalg.norm(centers_new-centers_old)  
```

```python
#plotting our final output after the algo
plt.scatter(X[:,0],X[:,1],s=50,c=cluster)
plt.scatter(centers_new[:,0],centers_new[:,1],marker='*', c='g', s=150)
```
